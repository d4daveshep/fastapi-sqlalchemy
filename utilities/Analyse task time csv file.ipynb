```python
df = pd.read_csv("./Utilisation report - 20230227.xlsx - TimeByTask.csv", usecols=[0,2,5,7,9], header=0)
df
```

```python
clients = list(df["[Job] Client"].loc[:])
clients
```

```python
client_counter = Counter(clients)
print(f"{len(set(client_counter))} clients")
```

```python
jobs = list(df["[Job] Name"].loc[:])
jobs
```

```python
job_counter = Counter(jobs)
print(f"{len(set(job_counter))} jobs")
```

```python
staff_counter = Counter(list(df["[Staff] Name"]))
staff_counter
```

```python
print(f"{len(set(staff_counter))} staff")
```

```python
nodes = set(client_counter).union(set(job_counter)).union(set(staff_counter))
print(f"{len(nodes)} nodes will be added")
```

```python
worked_on_connections = max([len(staff_counter),len(client_counter)])
worked_on_connections
```

```python
billed_time_connections = max([len(staff_counter),len(job_counter)])
billed_time_connections
```

```python
has_job_connections = max([len(client_counter),len(job_counter)])
has_job_connections
```

```python
total_connections = worked_on_connections + billed_time_connections + has_job_connections
total_connections
```
